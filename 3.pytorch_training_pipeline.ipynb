```python
import numpy as np
import pandas as pd
import torch
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
df.head()
```

```python
df.drop(columns=['id', 'Unnamed: 32'], inplace= True)
```

```python
df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, 1:], df.iloc[:, 0], test_size=0.2)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train.shape
```

```python
y_train
```

```python
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
```

```python
y_train
```

```python
x_train_tensor = torch.tensor(X_train,dtype=torch.float64,)
x_test_tensor = torch.tensor(X_test,dtype=torch.float64,)
y_train_tensor = torch.tensor(y_train,dtype=torch.float64,)
y_test_tensor = torch.tensor(y_test,dtype=torch.float64,)
```

```python
x_train_tensor
```

```python
"""defining the model"""
```

```python
class my_model():
  def __init__(self,x):
    self.wt = torch.rand(x.shape[1],dtype=torch.float64,requires_grad=True)
    self.bias = torch.zeros(1,dtype =torch.float64,requires_grad=True)

  def forward(self,x):
    z = torch.matmul(x,self.wt)+self.bias
    y_pred=torch.sigmoid_(z)
    return y_pred

  def loss_fun(self,y,y_pred):
    epsilon = 1e-7
    y_pred = torch.clamp(y_pred, epsilon, 1 - epsilon)
    loss = -(y*torch.log(y_pred)+(1-y)*torch.log(1-y_pred))
    return loss.mean()

```

```python
epochs = 25
lr = 0.001
```

```python
model = my_model(x_train_tensor)
model.wt.shape
for epoch in range(epochs):

  y_pred = model.forward(x_train_tensor)

  loss = model.loss_fun(y_train_tensor,y_pred)

  loss.backward()

  with torch.no_grad():
   model.wt -= lr*model.wt.grad
   model.bias-= lr*model.bias.grad

  model.wt.grad.zero_()
  model.bias.grad.zero_()

  print(f'epoch:{epoch+1},"  ",loss:{loss.item()}')


```

```python
# model evaluation
with torch.no_grad():
  y_pred = model.forward(x_test_tensor)
  y_pred = (y_pred > 0.5).float()
  accuracy = (y_pred == y_test_tensor).float().mean()
  print(f'Accuracy: {accuracy.item()}')
```
